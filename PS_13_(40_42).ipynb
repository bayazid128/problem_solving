```python
def abbrevName(name):
  return name.title()[0]+'.'+min(name.title()[name.index(' ')+1:])

print(abbrevName("Sam Harris"))
print(abbrevName("patrick feeney"))
```

```python
def check(a,x):
  return True if x in a else False

print(check('Hello World123','2'))
print(check('Hello World123','H'))
print(check('Hello World123','!'))
```
