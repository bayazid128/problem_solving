```python
def fake_bin(str):
  return ''.join('0' if i<'5' else '1' for i in str)
print(fake_bin('5632'))
```

```python
def sakura_fall(v):
  return 400/v if v>0 else 0

print(sakura_fall(5))
print(sakura_fall(0))
print(sakura_fall(-5))
```

```python
def replace_exclamation(sen):
  return ''.join('!' if i in 'aeiouAEIOU' else i for i in sen)

print(replace_exclamation("Hi!"))
print(replace_exclamation("!Hi! Hi!"))
print(replace_exclamation("aeiou"))
print(replace_exclamation("ABCDE"))
```
