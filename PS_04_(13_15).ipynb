```python
import math
def which_quarter(month):
  return "Quarter {}".format(math.ceil(month/3))   #(1,2,3) quarter 1,(4,5,6) quarter 2,(7,8,9) quarter 3,(10,11,12) quarter 4

print(which_quarter(int(input("which month: "))))
```
