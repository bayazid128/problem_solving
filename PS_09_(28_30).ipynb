```python
def find_average(ls):
  return 0 if len(ls)== 0 else sum(ls)/ len(ls)

print(find_average([100,99,99,98,98,98,97,97,97,97,5])) # 985 / 11 = 89.54545454545455
print(find_average([]))
```

```python
def mknegative(num):
  return -num if num>0 else num

print(mknegative(45))
print(mknegative(0))
print(mknegative(-980.01))
print(mknegative(float(input("Enter number: "))))
```

```python
def rps(p1,p2):
  if p1 == p2:
    return "Draw!"
  elif (p1=="rock" and p2=="scissors") or (p1=="scissors" and p2=="paper") or (p1=="paper" and p2=="rock"):
    return "Player 1 won!"
  else:
    return "Player 2 won!"

print(rps("scissors", "paper"))
print(rps("scissors", "rock"))
print(rps("paper", "paper"))
print(rps((input("Player 1: ")), (input("Player 2: "))))
```
