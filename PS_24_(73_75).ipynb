```python
def is_even(n):
  return n%2==0

print(is_even(float(input('Enter number: '))))
print(is_even(3))
print(is_even(2))
print(is_even(0))
print(is_even(-2))
print(is_even(-3))
```

```python
def better_than_average(peers_points,your_points):
  return (sum(peers_points)+your_points)/(len(peers_points)+1) < your_points

print(better_than_average([100,98,97],99))
print(better_than_average([80, 75, 90, 85],82.5))
```
