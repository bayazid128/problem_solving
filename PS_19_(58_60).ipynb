```python
def count_by(x,n):
  return [x*i for i in range(1,n+1)]

print(count_by(1,10))
print(count_by(2,5))
```

```python
def how_many_dalmatians(n):
  if n <= 10:
      return "Hardly any"
  elif n <= 50:
      return "More than a handful!"
  elif n < 101:
      return "Woah that's a lot of dogs!"
  else:
      return "101 DALMATIONS!!!"

print(how_many_dalmatians(int(input('Enter the number of dogs you have: '))))
```
