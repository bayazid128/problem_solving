```python
def sum_array(arr):
  arr.remove(max(arr)) if arr else arr
  arr.remove(min(arr)) if arr else arr

  return sum(arr) if arr else 0

print(sum_array([6, 2, 1, 8, 10]))
print(sum_array([1, 1, 11, 2, 3]))
print(sum_array(None))
print(sum_array([1]))
print(sum_array([1,2]))
```

```python
def combat(c_h, d_r): # c_h = 'current health' , d_r = 'damage recieved'
  return 'Your New Health = {}'.format(c_h - d_r if c_h>d_r else 0)

print(combat(200,90))
print(combat(150,150))
print(combat(150,200))
```

```python
def findSmallestInt(ls):
  return 'The smallest integer in this list is ({})'.format(min(ls))

print(findSmallestInt([34, 15, 88, 2]))
print(findSmallestInt([34, -345, -1, 100]))
```
