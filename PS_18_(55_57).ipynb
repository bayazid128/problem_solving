```python
def multi_table(num):
  return "\n".join(f"{i} * {num} = {i*num}" for i in range(1,11))

print(multi_table(int(input('NUM: '))))
```

```python
def bonus_time(salary, bonus):
  return str(f'${int(salary)*10 if bonus else int(salary)}')

print(bonus_time(50,True))
print(bonus_time(50,False))
print(type(bonus_time(50,True)))
```
