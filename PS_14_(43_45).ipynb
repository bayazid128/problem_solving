```python
def switch_it_up(n):
    match n:
        case 0:return'Zero'
        case 1:return'One'
        case 2:return'Two'
        case 3:return'Three'
        case 4:return'Four'
        case 5:return'Five'
        case 6:return'Six'
        case 7:return'Seven'
        case 8:return'Eight'
        case 9:return'Nine'

print(switch_it_up(2))
print(switch_it_up(20))
```

```python
def switch_it_up1(num):
  dict= {0:'Zero',1:"One",2:"Two",3:'Three',4:'Four',5:'Five',6:'Six',7:'Seven',8:'Eight',9:'Nine'}
  return dict.get(num)

print(switch_it_up1(0))
print(switch_it_up1(100))
```

```python
def lovefunc(num1,num2):
  return True if (num1%2==0 and num2%2!=0) or (num2%2==0 and num1%2!=0) else False

print(lovefunc(5,4))
print(lovefunc(4,5))
print(lovefunc(3,3))
print(lovefunc(int(input('Flower 1 petals: ')),int(input('Flower 2 petals: '))))
```
