```python
def count_sheep(n):
  sheep=""
  for i in range(1,n+1):
    sheep+= str(i)+ " sheep..."
  return sheep
print(count_sheep(3))
```

```python
def feast(beast,dish):
  return beast[0] == dish[0] and beast[-1] == dish[-1]

print(feast('great blue heron','garlic naan'))
print(feast('chickadee','chocolate cake'))
print(feast(input('Animal.s name: '),input('Dish: ')))
```
