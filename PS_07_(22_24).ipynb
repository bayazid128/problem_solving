```python
def is_uppercase(str):
  return True if str.isupper() else False

print(is_uppercase("HELLO WORLD!"))
print(is_uppercase("Hello World!"))
```

```python
def basic_op(operator, v1, v2):
  if operator == '+':
    return v1 + v2
  elif operator == '-':
    return v1 - v2
  elif operator == '*':
    return v1 * v2
  elif operator == '/':
    return v1 / v2

print(basic_op(input("operator: "),float(input("Value 1= ")),float(input("Value 2= "))))
```

```python
### In this function you can use math operations other than the four aforementioned basic operations.

def math_op(operator, v1, v2):
  return eval("{}{}{}".format(float(v1), operator, float(v2)))

print(math_op('%',115,20))
print(math_op('**',2,5))
print(math_op('>',100,50))
print(math_op('<',100,50))
```
