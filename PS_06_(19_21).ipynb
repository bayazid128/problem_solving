```python
def past(h,m,s):
  return "result = {}".format((h*3600+m*60+s) * 1000)

print(past(int(input("h = ")),int(input("m = ")),int(input("s = "))))
```

```python
def to_alternating_case(str):
  return str.swapcase()

print(to_alternating_case(input("Enter a string: ")))
```

```python
def zero_fuel(distance, mpg, gallons_left):
  return True if distance <= mpg * gallons_left else False

print(zero_fuel(50,25,2.5))
```

```python
### Another Solution with parameters as inputs
def zero_fuel(distance, mpg, gallons_left):
  return True if distance <= mpg * gallons_left else False

print(zero_fuel(float(input("Distance to the nearest pump= ")),float(input("Miles per gallon= ")),float(input("Gallons left= "))))
```
