```python
def square_sum(ls):
  return sum(i ** 2 for i in ls)

print(square_sum([1,2,2]))
print(square_sum([6,8]))
```

```python
def odd_count(num):
  return int(num/2) if num>0 else None

print(odd_count(7))
print(odd_count(15))
print(odd_count(int(input("Enter number: "))))
```

```python
def  real_floor(floor):
  if 0<floor< 13:
    return floor - 1
  elif floor>13:
    return floor - 2
  else:
    return floor

print(real_floor(1))
print(real_floor(5))
print(real_floor(16))
print(real_floor(int(input("Enter the floor number: "))))
```
