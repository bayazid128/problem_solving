```python
def twice_as_old(father_age, son_age):
  return abs(father_age - 2*son_age)

print(twice_as_old(13,8))
print(twice_as_old(20,10))
print(twice_as_old(20,5))
```

```python
def check_for_factor(base, factor):
  return base % factor==0

print(check_for_factor(0,10))
print(check_for_factor(7,2))
print(check_for_factor(6,3))
```
