```python
def swapValues(x,y):
  dict= {'num1':x,'num2':y}
  return "x= {}\ny= {}".format(dict.get('num2'), dict.get('num1'))

print(swapValues(float(input("x= ")),float(input("y= "))))
```

```python
def age_drink(age):
  if 0<age<14:
    return "drink toddy"
  elif age<18:
    return "drink coke"
  elif age<21:
    return "drink beer"
  elif age>20:
    return "drink whisky"

print(age_drink(float(input("Enter your age: "))))
```

```python
def hero(bullets,dragons):
  return bullets >= 2* dragons

print(hero(10,4))
print(hero(10,5))
print(hero(10,6))
print(hero(int(input("Enter number of bullets: ")),int(input("Enter number of dragons: "))))
```
