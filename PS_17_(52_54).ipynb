```python
def monkey_count(n):
  return [i for i in range(1,n+1)]

print(monkey_count(int(input('Number of monkeys: '))))
```

```python
def sum_str(num1,num2):
  return str(int(num1 or 0)+int(num2 or 0))

print(sum_str("4","5"))      # "4" + "5" --> "9"
print(sum_str("34","5"))     # "34" + "5" --> "39"
print(sum_str("",""))        # "" + "" --> "0"
print(sum_str("2",""))       # "2" + "" --> "2"
print(sum_str("-5","3"))     # "-5" + "3" --> "-2"
```
