```python
#Solution 1
def remove_char(str):
  length= len(str) - 1
  return str[1:length]

print (remove_char(input("Enter text here: ")))
```

```python
#Solution 2
def remove_chara(str):
  return str[1:-1]

print (remove_chara(input("Enter text here: ")))
```
